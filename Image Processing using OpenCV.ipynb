## OpenCV - Open Source Computer Vision

It is a library of pragramming functions mainly aimed at real time computer vision. It was created by Intel in 1999. It contains many popular algorithms for computer vision, including object detection and tracking algorithms built in.

You will observe an anomaly with the image displayed and that because of the way matplotlib and opencv deals with the three color channels (RGB).
 - Matplotlib: RED: GREEN: BLUE - RGB
 - OpenCV: BLUE: GREEN: RED - BGR

```python
# In order to display the image in matplotlib, we will have to convert the image from BGR (Which is the way OpenCV bring in the image) into RGB color scale
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
# How to load in a color image as grayscale using openCV
img_gray = cv2.imread(r'C:\Users\ikpro\OneDrive\Desktop\Deskstop\lion.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
```

```python
# Resizing Images
plt.imshow(img1)
```

```python
img1.shape
```

```python
# The Height and Width changes when using cv2.resize to Width and Height
plt.imshow(cv2.resize(img1, (2000, 800)))
```

```python
# Resize by ratios
width_ratio = 0.5
height_ratio = 0.5

r_img = cv2.resize(img1, (0,0), img1, width_ratio, height_ratio)
```

```python
plt.imshow(r_img)
```

```python
# Flip Images along the vertical and horizontal axis - 0 (Flip along the horizontal axis), 1 (Flip along the vertical axis), -1(Do both)
new_img = cv2.flip(img1, 0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(img1, -1)
plt.imshow(new_img)
```

```python
# Save a jpeg file
cv2.imwrite(r'C:\Users\ikpro\OneDrive\Desktop\Deskstop\u_lion.jpg', new_img)
```

### Drawing on Images

```python
# Lets create a blank image
b_img = np.zeros(shape=(512,512,3), dtype = np.int16)
b_img.shape
```

```python
# Display the image - Its black because we used np.zeroes and 0 means no color
plt.imshow(b_img)
```

```python
# We will start to draw some shapes on the blank image above - pt=(W, H) - Pt1(Top left corner) and Pt2 (Bottom right corner)
cv2.rectangle(b_img, pt1=(200,200), pt2=(300, 300), color=(255,0,0), thickness=5)
plt.imshow(b_img)
```

```python
# Drawing circles
cv2.circle(img=b_img, center=(100,100), radius=75, color=(255,10,150), thickness=5)
plt.imshow(b_img)
```

```python
# Fill the shape with a thick color

cv2.circle(img=b_img, center=(400,100), radius=50, color=(255,10,150), thickness=-1)
plt.imshow(b_img)
```

```python
cv2.line(b_img, pt1=(100,300), pt2=(200,400), color=(255,100,100))
plt.imshow(b_img)
```
